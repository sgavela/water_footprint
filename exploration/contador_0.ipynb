```python
import pandas as pd

from statsmodels.tsa.seasonal import seasonal_decompose

from pycaret import *
```

```python
df = pd.read_csv('../data/df6.csv')
df = df.drop(['YEAR_DAY','WEEKDAY','IS_GOOD'],axis=1)
#df.drop('DATE',axis=1,inplace=True)
```

```python
df0 = df[df['ID']==0]
ts0 = df0['DELTA']
ts0.index = pd.DatetimeIndex(df0['DATE'])
```

```python
result_add = seasonal_decompose(df0['DELTA'], model = 'add')
result_add.seasonal.plot()
```

```python
result_add.trend.plot()
```

```python
result_add.resid.plot()
```

```python
df0['DELTA_ROLL_7'] = df0['DELTA'].rolling(window=7).mean()
df0['DELTA_ROLL_14'] = df0['DELTA'].rolling(window=14).mean()
df0['DELTA_ROLL_28'] = df0['DELTA'].rolling(window=28).mean()
df0[['DELTA','DELTA_ROLL_7','DELTA_ROLL_14','DELTA_ROLL_28']].plot(figsize=(18,6))
```

```python
import pandas as pd
import pickle5 as pickle

df = pd.read_csv('../data/df6.csv')
df.head()
```

```python
df_all = df.drop(['ID','YEAR_DAY','WEEKDAY','IS_GOOD'],axis=1)
df_good = df[df['IS_GOOD']==1].drop(['YEAR_DAY','WEEKDAY','IS_GOOD'],axis=1)
```

```python
def get_temporal_serie(i, df=df_good):
    df_aux = df[df['ID']==0]
    df_aux = df_aux['DATE']
    df_aux = df_aux.drop(['DATE','ID','MEAN_CONSUMPTION','VARIANCE_CONSUMPTION'],axis=1)
    return df_aux
```

```python
df_all['LAG_1'] = df_all['DELTA'].shift(1, fill_value=0)
df_all['LAG_2'] = df_all['DELTA'].shift(2, fill_value=0)
df_all['LAG_3'] = df_all['DELTA'].shift(3, fill_value=0)
df_all['LAG_4'] = df_all['DELTA'].shift(4, fill_value=0)
df_all['LAG_5'] = df_all['DELTA'].shift(5, fill_value=0)
df_all['LAG_6'] = df_all['DELTA'].shift(6, fill_value=0)
df_all['LAG_7'] = df_all['DELTA'].shift(7, fill_value=0)
df_all['LAG_14'] = df_all['DELTA'].shift(14, fill_value=0)
```

```python
df_all.drop('DATE',axis=1,inplace=True)
```

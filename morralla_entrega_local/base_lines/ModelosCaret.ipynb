```python
import pandas as pd
```

```python
datos = pd.read_csv('enriquecidos-004.pkl')
datos
```

```python
datos.columns
```

```python
datos.drop(columns = ['Unnamed: 0', 'DATE'], inplace = True)
```

```python
groups = datos.groupby('ID').sum()
groups['TotalCons'] = groups['DELTA']
groups = groups['TotalCons']
```

```python
groups2 = datos.groupby('ID').var()
groups2['Variance'] = groups2['DELTA']
groups2 = groups2['Variance']
```

```python
datos = pd.merge(datos, groups, left_on='ID', right_on = 'ID')
datos = pd.merge(datos, groups2, left_on='ID', right_on = 'ID')
datos.astype({"ID":'category'})
```

```python
from pycaret.regression import *
datos['DELTA'] = datos['DELTA'].apply(lambda x: float(x))
init = setup(datos, target = 'DELTA', train_size = 0.9, use_gpu = True, n_jobs = 16)
```

```python
best = compare_models(budget_time = 100)
```

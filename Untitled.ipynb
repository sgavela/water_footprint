```python
import pandas as pd
```

```python
results_good_counters_xgb = pd.read_pickle('./data/results_good_counters_xgb.pkl')
results_good_counters_xgb.shape[0]/14
```

```python
results_bad_counters_xgb = pd.read_pickle('./data/results_bad_counters_xgb.pkl')
results_bad_counters_xgb.shape[0]/14
```

```python
results_good_counters_prophet = pd.read_pickle('./data/results_prophet_february.pkl')
results_good_counters_prophet.head()
```

## Entrega 1 (xgb para todo)

```python
all_counters_results_1 = pd.concat([results_good_counters_xgb, results_bad_counters_xgb])
all_counters_results_1 = all_counters_results_1.sort_values(['ID','DATE'])
```

```python
all_counters_results_1['predictions'][all_counters_results_1['predictions'] < 0] = 0
```

```python
final_df_1 = pd.DataFrame(index = all_counters_results_1['ID'].unique())
```

```python
test_dates = ['2020-02-01','2020-02-02','2020-02-03','2020-02-04','2020-02-05','2020-02-06','2020-02-07',
              '2020-02-08','2020-02-09','2020-02-10','2020-02-11','2020-02-12','2020-02-13','2020-02-14']
for date in test_dates:
    final_df_1[date] = all_counters_results_1[all_counters_results_1['DATE']==date]['predictions'].values
```

```python
final_df_1['Semana_1'] = final_df_1[['2020-02-01','2020-02-02','2020-02-03','2020-02-04','2020-02-05','2020-02-06','2020-02-07']].sum(axis=1)
final_df_1['Semana_2'] = final_df_1[['2020-02-08','2020-02-09','2020-02-10','2020-02-11','2020-02-12','2020-02-13','2020-02-14']].sum(axis=1)
final_df_1.drop(['2020-02-08','2020-02-09','2020-02-10','2020-02-11','2020-02-12','2020-02-13','2020-02-14'], axis=1, inplace=True)
```

```python
final_df_1['ID'] = final_df_1.index
final_df_1.head()
```

```python
final_df_1 = final_df_1[['ID','2020-02-01','2020-02-02','2020-02-03','2020-02-04','2020-02-05','2020-02-06','2020-02-07','Semana_1','Semana_2']]
final_df_1.head()
```

```python
final_df_1.shape
```

```python
final_df_1 = round(final_df_1,2)
```

```python
round(final_df_1['2020-02-02'],3)
```

```python
round(final_df_1,2)
```

```python
final_df_1.shape
```

```python
path = './data/Cajamar_Analistas Financieros Internacionales Escuela (AFI)_La Manzana de Newton_1.txt'
final_df_1.to_csv(path, sep = '|', header=False, index=False)
```
